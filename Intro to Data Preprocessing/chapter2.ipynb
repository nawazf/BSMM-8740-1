```python
import pandas as pd
import numpy as np
```

```python
wine = pd.read_csv('../Intro to Data Preprocessing/Datasets/wine_types.csv')
```

```python
wine.head()
```

```python
wine.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
wine_scaled = wine[['Ash', 'Alcalinity of ash', 'Magnesium']]
```

```python
wine_subset_scaled = scaler.fit_transform(wine_scaled)
```

```python
print(wine_subset_scaled.var())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine_subset_scaled, wine['Type'], test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

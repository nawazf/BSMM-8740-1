```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
dataset.info()
```

```python
numeric_data = dataset.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
numeric_data  = ['CreditScore',
 'Age',
 'Tenure',
 'Balance',
 'NumOfProducts',
 'EstimatedSalary'
]
```

```python
numeric_data
```

```python
categorialdata = dataset.select_dtypes(include=['category']).columns.tolist()
```

```python
categorialdata
```

```python
target = ['Exited']
```

```python
binary_data = [ 'HasCrCard',
 'IsActiveMember']
```

```python
dataset[target]
```

```python
X = dataset[numeric_data + categorialdata + binary_data]
```

```python
X
```

```python
y = dataset[target]
```

```python
y
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='median'), numeric_data),
        ('cat', OneHotEncoder(), categorialdata)]
)
```

```python
preprocessor
```

```python
type(preprocessor)
```

```python
X 
```

```python
X = preprocessor.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=43)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_cl = DecisionTreeClassifier(random_state=42)
```

```python
dt_cl.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
y_test_pred = dt_cl.predict(X_test)
print(f'Area Under the ROC Curve: {roc_auc_score(y_test, y_test_pred): 0.4f}')
print(f'Accuracy: {accuracy_score(y_test, y_test_pred): 0.4f}')
print(f'F1 Score: {f1_score(y_test, y_test_pred): 0.4f}')
print(classification_report(y_test, y_test_pred, digits=4))
pd.DataFrame(confusion_matrix(y_test, y_test_pred))
```

```python
from sklearn.model_selection import GridSearchCV
clf_df = DecisionTreeClassifier(class_weight='balanced', random_state=42)
param_grid_dt = {'criterion': ['gini', 'entropy'],
                 'max_depth': list(range(4, 250))}
grid_dt = GridSearchCV(clf_df, param_grid_dt, cv=5, scoring='roc_auc', n_jobs=-1, verbose=20)
grid_dt.fit(X_train, y_train)

print(f'Best Score: {grid_dt.best_score_}')
print(f'Best Parameters: {grid_dt.best_params_}')
print(f'Best Estimator: {grid_dt.best_estimator_}')
```

```python
y_test_pred = grid_dt.predict(X_test)
print(f'Area Under the ROC Curve: {roc_auc_score(y_test, y_test_pred): 0.4f}')
print(f'Accuracy: {accuracy_score(y_test, y_test_pred): 0.4f}')
print(f'F1 Score: {f1_score(y_test, y_test_pred): 0.4f}')
print(classification_report(y_test, y_test_pred, digits=4))
pd.DataFrame(confusion_matrix(y_test, y_test_pred))
```

```python
dataset = pd.read_csv('Car_Insurance_Claim.csv')
dataset.info()
```

```python
dataset= dataset.set_index('ID')
```

```python
object_columns = dataset.select_dtypes(include=['object']).columns
object_columns_data = dataset[object_columns]

object_columns_data
```

```python
numerical_columns = dataset.select_dtypes(include=['int64','float64']).columns
numerical_columns_data = dataset[numerical_columns]
numerical_columns_data.describe()
```

```python
categorical_features = ['AGE', 'GENDER', 'RACE', 'DRIVING_EXPERIENCE', 
                        'EDUCATION', 'INCOME', 'VEHICLE_OWNERSHIP', 
                        'VEHICLE_YEAR', 'MARRIED', 'CHILDREN', 
                        'POSTAL_CODE', 'VEHICLE_TYPE']
numrerical_features = ['CREDIT_SCORE', 'ANNUAL_MILEAGE', 'SPEEDING_VIOLATIONS', 
                    'DUIS', 'PAST_ACCIDENTS']
target = ['OUTCOME']
```

```python
X = dataset[categorical_features + numrerical_features]
y = dataset[target]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numrerical_features),
        ('cat', OneHotEncoder(), categorical_features)])
```

```python
X = preprocessor.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train.values.ravel())
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
print('Accuracy score: ', accuracy_score(y_test, y_pred))
print('Classification report: ', classification_report(y_test, y_pred))
print('ROC AUC: ', roc_auc_score(y_test, y_pred))
print('AUC Score ', roc_curve(y_test, y_pred))
pd.DataFrame(roc_curve(y_test, y_pred),
             index=['False Positive Rate', 'True Positive Rate', 'Thresholds'], columns=['Value1', 'Value2', 'Value3'])
```

```python
print('Confusion matrix: ', confusion_matrix(y_test, y_pred))
pd.DataFrame(confusion_matrix(y_test, y_pred),
      index = ['Actual Negative', 'Actual Positive'],columns = ['Predicted Negative', 'Predicted Positive'])
```

```python
trian_accuracy = []
test_accuracy = []
```

```python
negigbors_settings = range(1, 11)
for neighbor in negigbors_settings:
    clf = KNeighborsClassifier(n_neighbors=neighbor)
    clf.fit(X_train, y_train.values.ravel())
    trian_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(12,8))
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(negigbors_settings, trian_accuracy, label='Training Accuracy')
plt.plot(negigbors_settings, test_accuracy, label='Test Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

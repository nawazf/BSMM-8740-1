```python
# Tokenize the article: tokens
tokens = word_tokenize(sentence)

# Convert the tokens into lowercase: lower_tokens
lower_tokens = [t.lower() for t in tokens]

# Create a Counter with the lowercase tokens: bow_simple
bow_simple = Counter(lower_tokens)

# Print the 10 most common tokens
print(bow_simple.most_common(5))
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# remove stop words
no_stops = [t for t in tokens
            if t not in stopwords.words('english')]

no_stops
```

```python
Counter(no_stops)
```

```python
Counter(no_stops).most_common(2)
```

```python
# Import WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

english_stops = stopwords.words('english')

# Retain alphabetic words: alpha_only
alpha_only = [t for t in lower_tokens if t.isalpha()]

# Remove all stop words: no_stops
no_stops = [t for t in alpha_only if t not in english_stops]

# Instantiate the WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Lemmatize all tokens into a new list: lemmatized
lemmatized = [wordnet_lemmatizer.lemmatize(t) for t in no_stops]

# Create the bag-of-words: bow
bow = Counter(lemmatized)

# Print the 10 most common tokens
print(bow.most_common(10))
```

## Introduction to Gensim

```python
from gensim.corpora.dictionary import Dictionary
from nltk.tokenize import word_tokenize
```

```python
my_documents = ['This movie was about a spaceship and aliens.',
                'I really liked the movie!',
                'Awesome action scenes, but boring characters.',
                'This movie was awful! I hate alien films.',
                'Space is cool! I liked the movie.',
                'More space films, please!']
my_documents
```

```python
tokenized_docs = [word_tokenize(doc.lower())
                  for doc in my_documents  ]
tokenized_docs
```

```python
len(tokenized_docs)
```

```python
len(tokenized_docs[0])
```

```python
# Dictionary of all tokens and their IDs
dictionary = Dictionary(tokenized_docs)
dictionary.token2id
```

```python
# this generat4e a list of lists
# each list represents one document
# each document is a series of tuples (token id , frequency)

corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]
corpus
```

### TF-IDF with gensim

```python
from gensim.models.tfidfmodel import TfidfModel

# for our second document, we see the token weight along with the token ids
# These weights can help you determine keywords and topics with corpus with shared vocabulary
tfidf = TfidfModel(corpus)
tfidf[corpus[1]]
```

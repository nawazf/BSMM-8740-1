```python
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
display(Markdown("### Rows and columns"))

print(X.shape)
print(y.shape)

print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(X_train, y_train)
```

```python
y_predicted_test = model_knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
display(Markdown("#### Accuracy and Confusion Matrix"))
print()
print(f'Test Accuracy {accuracy_score(y_test, y_predicted_test): 0.4f}')
print()
pd.DataFrame(confusion_matrix(y_test, y_predicted_test),
             index=['Actual 0', 'Actual 1', 'Actual 2'], columns=['Predicted 0', 'Predicted 1', 'Predicted 2'])
```
